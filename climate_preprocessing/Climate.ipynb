```python
import numpy as np
import pandas as pd
```

```python
temp2014= pd.read_csv('Climate_2014.csv')
temp2015= pd.read_csv('Climate_2015.csv')
temp2016= pd.read_csv('Climate_2016.csv')
temp2017= pd.read_csv('Climate_2017.csv')
temp2018= pd.read_csv('Climate_2018.csv')
temp2019= pd.read_csv('Climate_2019.csv')
temp2019.head()
```

```python
temp_df = pd.concat([temp2014,temp2015,temp2016,temp2017,temp2018,temp2019], axis=0)
temp_df.shape
```

```python
temp_df['Max Temp (°C)'].isna().sum()
```

```python
temp_df = temp_df[['Date/Time', 'Mean Temp (°C)', 'Total Rain (mm)', 'Total Snow (cm)', 'Total Precip (mm)', 'Snow on Grnd (cm)']]
temp_df.head()
```

```python
temp_df = temp_df.rename(columns={'Mean Temp (°C)': 'mean_temp', 
                                  'Total Rain (mm)': 'total_rain',
                                  'Total Snow (cm)': 'total_snow',
                                  'Total Precip (mm)': 'total_precip',
                                  'Snow on Grnd (cm)': 'snow_on_ground' })
temp_df.head()
```

```python
temp_df.to_csv("torontotemp_2014_to_2019.csv", index=False)
```
